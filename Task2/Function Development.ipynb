## NumPy Function

```python
# Main Code
import numpy as np
import time
```

#### Creating the get_array_stats function:

```python
# Beginning of Function
def get_array_stats(arr: np.ndarray[np.int64]): # Parameter is an Array of type NumPy Array
# only accepts 64bit integer arrays
    start_t = time.time() # Records Starting Time
    column_mean = np.mean(arr, axis = 0) # Calculates Means along 0 axis (Columns)
    column_std = np.std(arr, axis = 0) # Calculates Standard Deviation along 0 axis (Columns)
    array_min = np.amin(arr) # Finds he smallest number within the entire array
    array_max = np.amax(arr) # Finds the largest number within the entire array
    end_t = time.time() # Recrods End Time
    computation_time = end_t - start_t # Calculates time taken for computation
    return column_mean, column_std, array_min, array_max, computation_time # Returns all values packed in a tuple
```

#### Testing:

```python
# Test Function Which Has parameters of Size
def test_function(row,col) -> None:
    random_array = np.random.randint(low=0, high=100000, size=(row,col))
    print(random_array)
    print()
    print("1. Mean (column-wise) 2. Standard deviation (column-wise) 3. Minimum value 4. Maximum value 5. Computation time")
    print()
    print(get_array_stats(random_array))

# Test for 10 x 10, 100 x 100, 100 x 1000
test_function(10,10)
print("-------")
test_function(100,100)
print("-------")
test_function(100,1000)

```
